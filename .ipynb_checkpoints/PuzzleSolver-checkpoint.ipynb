```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import cv2
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression

def analyzePicture(picNo):
    path = "img/teile_second/"
    if picNo < 10:
        path += "000"
    elif picNo < 100:
        path += "00"
    elif picNo < 1000:
        path += "0"
    else:
        path = path
    path += str(picNo) + ".jpg"
    
    img = cv2.imread(path)
    h,w = img.shape[0:2]
    #base[5:h+5,5:w+5]=img
    #img = base
    img2 = np.zeros((h+6,w+6, 3), dtype=np.uint8)
    img2 = cv2.bitwise_not(img2)
    img2[3:h+3,3:w+3] = img
    
    gray   = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
    h,w = gray.shape[0:2]
    base_size = h+2, w+2
    base = np.zeros(base_size, dtype=np.uint8)
    thresh = cv2.threshold(gray, 240, 255, cv2.THRESH_BINARY)[1]
    
    cv2.floodFill(thresh, base, (0,0), 255)
    thresh = base
    
    plt.imshow(thresh)
    plt.show()
    contours, hierarchy = cv2.findContours(thresh,cv2.RETR_CCOMP, cv2.CHAIN_APPROX_NONE) 
    
    h,w = thresh.shape[0:2]
    img2 = np.zeros((h+6,w+6, 3), dtype=np.uint8)
    img2 = cv2.bitwise_not(img2)
    img = cv2.imread(path)
    h,w = img.shape[0:2]
    img2[4:h+4,4:w+4] = img
    image = img2.copy()
    cont = len(contours)-1
    cv2.drawContours(img2, contours, cont , (0, 0, 255), 1) 
    plt.imshow(img2)
    plt.show()
    
    corners = cv2.goodFeaturesToTrack(thresh,30,0.1,10)
    corners = np.int0(corners)
    
    imX, imY, imZ= image.shape
    points = []
    c1, c2, c3, c4 = [], [], [], []
        
    for i in range(0,len(corners)):
        x,y = corners[i].ravel()      
        points.append([i,(x,y)])
        c1.append([i,np.linalg.norm((x,y))])
        c2.append([i,np.linalg.norm((x-imX,y))])
        c3.append([i,np.linalg.norm((x-imX,y-imY))])
        c4.append([i,np.linalg.norm((x,y-imY))])
        
    def sort(e):
        return e[len(e)-1]
    
    maxCorn = 4
    
    c1.sort(key=sort)
    c2.sort(key=sort)
    c3.sort(key=sort)
    c4.sort(key=sort)
    for x in range(0,maxCorn):
        cv2.circle(image,points[c1[x][0]][1],2,(255,0,0),-1)
        cv2.circle(image,points[c2[x][0]][1],2,(0,255,0),-1)
        cv2.circle(image,points[c3[x][0]][1],2,(255,0,0),-1)
        cv2.circle(image,points[c4[x][0]][1],2,(0,255,0),-1)
    
    calcArray = []
    
    def totalScalarProducts(a,b,c,d):
        
        ab = np.subtract(b,a)
        bc = np.subtract(c,b)
        cd = np.subtract(d,c)
        da = np.subtract(a,d)
        
        c1 = np.abs(np.dot(ab,da))
        c2 = np.abs(np.dot(ab,bc))
        c3 = np.abs(np.dot(bc,cd))
        c4 = np.abs(np.dot(cd,da))      
        x = c1+c2+c3+c4
        
        return x
    
    def totalArea(a,b,c,d):
        
        ab = np.linalg.norm(np.subtract(b,a))
        bc = np.linalg.norm(np.subtract(c,b))
        ca = np.linalg.norm(np.subtract(a,c))
        cd = np.linalg.norm(np.subtract(d,c))
        da = np.linalg.norm(np.subtract(a,d))
        
        s1 = (ab + bc + ca) / 2
        s2 = (cd + da + ca) / 2
        
        a1 = (s1*(s1-ab)*(s1-bc)*(s1-ca)) / 2
        a2 = (s2*(s2-cd)*(s2-da)*(s2-ca)) / 2
        a = a1+a2
        
        return a
        
    
    for i in range(0,maxCorn):
        for j in range(0,maxCorn):
            for k in range(0,maxCorn):
                for l in range(0,maxCorn):
                    ai = c1[i][0]
                    bi = c2[j][0]
                    ci = c3[k][0]
                    di = c4[l][0]
                    a = list(points[ai][1])
                    b = list(points[bi][1])
                    c = list(points[ci][1])
                    d = list(points[di][1])
                    if a!=b and b!=c and c!=d and d!=a:
                        calcArray.append([ai,bi,ci,di,totalScalarProducts(a,b,c,d), totalArea(a,b,c,d)])
    def sortArray(e):
        return 1/(e[5])
    
    calcArray.sort(key=sortArray)
    
    p1 = points[calcArray[0][0]][1]
    p2 = points[calcArray[0][1]][1]
    p3 = points[calcArray[0][2]][1]
    p4 = points[calcArray[1][3]][1]
    cv2.circle(image,p1,5,(0,0,255),-1)
    cv2.circle(image,p2,5,(0,0,255),-1)
    cv2.circle(image,p3,5,(0,0,255),-1)
    cv2.circle(image,p4,5,(0,0,255),-1)
   
   # converting contour to array of points 
    edge = []
    for x in range(1,len(contours)):
        for p in contours[x]:
            point = list(p[0])
            point = (point[0],point[1])
            edge.append(point)
    
    edge1= []
    edge2= []
    edge3= []
    edge4= []
    maxDist = 25
    
    for i in range(0,len(edge)):
        if np.linalg.norm(np.subtract(edge[i],p1))<maxDist:
            edge1.append([edge[i],i])
        elif np.linalg.norm(np.subtract(edge[i],p2))<maxDist:
            edge2.append([edge[i],i])
        elif np.linalg.norm(np.subtract(edge[i],p3))<maxDist:
            edge3.append([edge[i],i])
        elif np.linalg.norm(np.subtract(edge[i],p4))<maxDist:
            edge4.append([edge[i],i])
     
    
    def getRscore(edge):
        x_data = []
        y_data = []
        #for x in range(int(len(edge)/2),len(edge)):
        for x in range(0,int(len(edge)/2)):
            x_data.append(edge[x][0][0])
            y_data.append(edge[x][0][1])
            cv2.circle(image,edge[x][0],1,(0,0,255),-1)
    getRscore(edge3)
    
    
    plt.imshow(image)
    plt.show
    cv2.imwrite("img/tempCornImg/"+str(picNo)+".jpg", image)
```
